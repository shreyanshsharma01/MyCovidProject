Detailed overview of COVID 19 in Delhi

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('Delhi data june-july.csv')
```

```python
df.info()
```

```python
df.head(45)
```

```python
#Plot of Daywise Analysis
plt.figure(figsize=(15, 6))
t=plt.title('Datewise Analysis of Delhi : ')
t.set_position([.4, 1.04])
plt.plot(df.Date,df.Confirmed,color='red',label='Confirmed Cases', marker='o')
plt.plot(df.Date,df.Recovered,color='limegreen',label='Recovered Cases', marker='o')
plt.plot(df.Date,df.Deceased,color='brown',label='Deceased Cases', marker='o')

plt.xlabel('Date:1 June to 15 July',labelpad=10)
plt.ylabel('Cases',labelpad=30,rotation=0)
plt.xticks(rotation=45)
plt.savefig('Datewise analysis of cases.png')
plt.legend()
plt.show()
```

```python
#Plotting as bargraph
plt.figure(figsize=(10, 6))
plt.bar(df.Date, df.Confirmed,color='tomato', label='Confirmed Cases')
plt.bar(df.Date, df.Recovered,color='turquoise', label='Recovered Cases')
plt.bar(df.Date, df.Deceased,color='darkslategray', label='Deceased Cases')
plt.title('Datewise Analysis of Delhi: ')
plt.xticks(rotation=90)
plt.legend()
plt.xlabel('Date:1 June to 15 July',labelpad=10)
plt.ylabel('Number of Cases',labelpad=40,rotation=0)
plt.show()
plt.savefig('Datewise Analysis of Delhi bargraph.png')
```

```python
#piechart
labels=['Confirmed','Deceased','Recovered']
sizes=[97149,3014,87221]
colors = ['tomato', 'darkslategray', 'turquoise']
explode = (0, 0.2, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',textprops={'color':"black",'fontsize':13},shadow=True, center=(0.5,0.5), startangle=0)
plt.axis('equal')
plt.show()
plt.savefig('piechart1.png')
```

```python
states_n=['Rajasthan', 'Delhi', 'U.P', 'Haryana']
x_index=np.arange(4)
width=0.25# the width of the bars
conf=[32334,126323,55588,28186]
plt.bar(x_index-width, conf, color='tomato', width=width, label='Confirmed')
recv=[23364,107650,33500,21697]
plt.bar(x_index, recv, color='#fae987', width=width, label='Recovered')
dth=[583,3719,1263,372]
plt.bar(x_index+width, dth,color='#14213d', width=width, label='Deceased')
plt.legend()
plt.title('Latest Statewise covid19 Status')
plt.xlabel('States')
plt.ylabel('Cases',rotation=0)
plt.xticks(x_index+width/2, states_n)
plt.show()
plt.savefig('Latest Statewise covid19.png')

```

```python
#data of confirmed cases monthly
Dates=['15-06','20-06','25-06','30-06','05-07','10-07','15-07']
RJ=[287,381,287,348,632,611,866]
DL=[1647,3630,3390,2199,2244,2089,1647]
UP=[476,541,636,664,1153,1338,1659]
HR=[514,480,453,338,457,570,678]
x=np.arange(len(Dates))

#plot data
#use zorder to put bar in front of the grid
bar_width=0.2
plt.bar(x-bar_width, RJ, width=bar_width, color='#9D0208',zorder=2,label='RJ')
plt.bar(x, DL, width=bar_width, color='#DC2F02',zorder=2,label='DL')
plt.bar(x+bar_width, UP, width=bar_width, color='#E85D04',zorder=2,label='UP')
plt.bar(x+bar_width*2, HR, width=bar_width, color='#F48C06',zorder=2,label='Haryana')
plt.legend()
plt.title('Monthly Confirmed cases Statewise Status')
plt.xlabel('Date:15 June to 15 July')
plt.ylabel('Cases',rotation=0,labelpad=25)
plt.xticks(x+bar_width/2, Dates)
plt.grid(axis='y')#grid
plt.show()
plt.savefig('Monthly Confirmed cases.png')
```

```python
#data of recovered cases monthly
Dates=['15-06','20-06','25-06','30-06','05-07','10-07','15-07']
RJ=[219,277,229,299,288,550,333]
DL=[604,7725,3328,2113,3083,2468,2463]
UP=[342,374,533,578,607,660,762]
HR=[562,239,455,470,687,399,577]
x=np.arange(len(Dates))

#plot data
#use zorder to put bar in front of the grid
bar_width=0.2
plt.bar(x-bar_width, RJ, width=bar_width, color='#05668d',zorder=2,label='RJ')
plt.bar(x, DL, width=bar_width, color='#00a896',zorder=2,label='DL')
plt.bar(x+bar_width, UP, width=bar_width, color='#02deaf',zorder=2,label='UP')
plt.bar(x+bar_width*2, HR, width=bar_width, color='#73e7ce',zorder=2,label='Haryana')
plt.legend()
plt.title('Monthly Recovered cases Statewise Status')
plt.xlabel('Date:15 June to 15 July')
plt.ylabel('Cases',rotation=0,labelpad=25)
plt.xticks(x+bar_width/2, Dates)
plt.grid(axis='y')#grid
plt.show()
plt.savefig('Monthly Recovered cases.png')

```

```python
#CMR(Crude Mortality Rate)
#RCR(Recovered to Confirmed Ratio)
states_n= ['Rajasthan','Delhi','UP', 'Haryana']
RCR= [.643,1.293,.596,.97]
CMR = [0.0095,0.296,0.0085,0.0191]

x = np.arange(len(states_n))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, RCR,color='tomato',width=width,label='RCR:Recovered to Confirmed Ratio')
rects2 = ax.bar(x + width, CMR,color='#52708e',width=width, label='CMR(per 100,000 person)')

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.legend()
plt.title('RCR vs CMR statewise')
plt.xlabel('States')
plt.ylabel('RCR AND CMR (in %)')
plt.xticks(x+width/2, states_n)

def autolabel(rects):
    #Attach a text label above each bar in *rects*, displaying its height
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width()/2  , height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
fig.tight_layout()
plt.show()
plt.savefig('RCR vs CMR statewise.png')
```
